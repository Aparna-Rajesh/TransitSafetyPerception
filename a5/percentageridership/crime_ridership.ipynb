```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CHECK FOR EXACT MONTH-TO-MONTH MATCH BETWEEN RIDERSHIP AND CRIME DATASETS

data1 = pd.read_csv('../crime/allcrime.csv')  # Change 'file1.csv' to your first file
data2 = pd.read_csv('../ridership/allridership.csv')  # Change 'file2.csv' to your second file

data1['month'] = pd.to_datetime(data1['month']).dt.to_period('M')
data2['month'] = pd.to_datetime(data2['month']).dt.to_period('M')

months_in_data1 = set(data1['month'])
months_in_data2 = set(data2['month'])

all_exist = months_in_data1.issubset(months_in_data2)
print("All months in the first file exist in the second file:", all_exist)
```

```python
# DATA CHECK FOR PERCENTAGE...raw graph just to see if the numbers make sense

# Load the datasets
data1 = pd.read_csv('../crime/recentcrime.csv')  # Data for crimes
data2 = pd.read_csv('../ridership/recentridership.csv')  # Data for ridership

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Plotting Crime Data
axes[0].plot(data1['month'], data1['total'], marker='o', linestyle='-')
axes[0].set_title('Crime Count Over Time')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Crime Count')
january_ticks_crime = [month if month.endswith('-01') else '' for month in data1['month']]
axes[0].set_xticks(data1['month'])
axes[0].set_xticklabels(january_ticks_crime, rotation=45)
axes[0].grid(True)

# Plotting Ridership Data
axes[1].plot(data2['month'], data2['ridership'], marker='o', color='r', linestyle='-')
axes[1].set_title('Ridership Over Time')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Ridership Count')
january_ticks_ridership = [month if month.endswith('-01') else '' for month in data2['month']]
axes[1].set_xticks(data2['month'])
axes[1].set_xticklabels(january_ticks_ridership, rotation=45)
axes[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
