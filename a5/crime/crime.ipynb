```python
import pandas as pd
```

```python
# CHOP IT TO MATCH RIDERSHIP DATA (2016-01 AND ONWARD)

data = pd.read_csv('crime2015-2019.csv')
data['month'] = pd.to_datetime(data['month']).dt.to_period('M')
filtered_data = data[data['month'] >= '2016-01']
filtered_data.to_csv('output_filtered.csv', index=False)
```

```python
# FIND THE HIGHEST DATE SO WE CAN CUT THE CRIME DATASET DOWN

data = pd.read_csv('crime2015-2019.csv')
data['month'] = pd.to_datetime(data['month'])
max_date = data['month'].max()
print(max_date)
```

```python
# SHITTY VISUALIZATION (recent)

import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('recentcrime.csv')

# Plotting
plt.plot(data['month'], data['total'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Crime Count Over Time')

# Customize x-axis ticks to include only January labels
january_ticks = [month if month.endswith('-01') else '' for month in data['month']]
plt.xticks(data['month'], january_ticks, rotation=45)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
