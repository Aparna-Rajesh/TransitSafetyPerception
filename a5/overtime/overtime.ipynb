```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# MAKE THE SHITTY GRAHIC

# Read data from CSV
df = pd.read_csv('overtime.csv')

# Convert 'Total OT Pay' from string to float for plotting
df['Total OT Pay'] = df['Total OT Pay'].replace('[\$,]', '', regex=True).astype(float)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Total OT Pay'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Total OT Pay ($)')
plt.title('Total OT Pay by Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
